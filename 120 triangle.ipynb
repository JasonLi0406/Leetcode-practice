```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """
        n = len(triangle)

        dp = [[0 for _ in range(n)] for _ in range(2)]

        dp[0][0] = triangle[0][0]

        for row in range(1, n):
            for col in range(row+1):

                if col == 0:
                    dp[1][0] = triangle[row][0] + dp[0][0]

                elif col == row:
                    dp[1][col] = triangle[row][col] + dp[0][col-1]
                
                else:
                    dp[1][col] = triangle[row][col] + min(dp[0][col-1], dp[0][col])
            
            dp[0] = dp[1][:]
        
        return min(dp[0])
```

```python
def minimumTotal(triangle):
    """
    :type triangle: List[List[int]]
    :rtype: int
    """
    n = len(triangle)

    dp = [0 for _ in range(n)]

    dp[0] = triangle[0][0]

    for row in range(1, n):
        for col in range(row, -1, -1):

            if col == 0:
                dp[0] += triangle[row][0]

            elif col == row:
                dp[col] = triangle[row][col] + dp[col-1]
            
            else:
                dp[col] = triangle[row][col] + min(dp[col-1], dp[col])
    
    return min(dp)
```

```python
def minimumTotal(triangle):

    dp = triangle[-1][:]

    for i in range(len(triangle)-2, -1, -1):
        for j in range(i+1):
            dp[j] = triangle[i][j] + min(dp[j], dp[j+1])
            
    return dp[0]
```

```python
triangle = [[-1],[3,2],[-3,1,-1]]

minimumTotal(triangle)
```

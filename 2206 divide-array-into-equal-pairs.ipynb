```python
class Solution(object):
    def divideArray(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        from collections import Counter

        counts = Counter(nums)

        for i in counts.values():
            if i%2 == 1:
                return False
        return True
```

```python
def divideArray(nums):
    """
    :type nums: List[int]
    :rtype: bool
    """
    nums.sort()

    for i in range(0, len(nums), 2):
        if nums[i] != nums[i+1]:
            return False
    return True
```

```python
nums = [4,3,2,1]
divideArray(nums)
```

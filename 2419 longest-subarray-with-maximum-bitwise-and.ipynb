```python
class Solution:
    def longestSubarray(self, nums: list[int]) -> int:
        Max, cnt = nums[0], 0
        res = 0
        for i in range(len(nums)):
            if nums[i] > Max:
                Max = nums[i]
                res = cnt = 1

            elif nums[i] == Max:
                cnt += 1
                res = max(res, cnt)
            
            else:
                cnt = 0

        return res
```

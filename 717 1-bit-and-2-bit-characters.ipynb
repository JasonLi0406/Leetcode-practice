```python
from typing import List
```

```python
class Solution:
    def isOneBitCharacter(self, bits: List[int]) -> bool:
        n = len(bits)
    
        i = 0   
        bit = bool(bits[0])
        while i < n:
            bit = bool(bits[i])
            if bit:
                i += 2
            else:
                i += 1
        
        return not(bit)
```

```python
from typing import List
class Solution:
    def minCost(self, colors: str, neededTime: List[int]) -> int:
        n = len(colors)

        # Initiate the dp array
        dp = [0] * n
        dp[0] = neededTime[0]

        # Main loop
        res = 0
        for i in range(1, n):
            if colors[i-1] == colors[i]:
                res += min(dp[i-1], neededTime[i])
                dp[i] = max(dp[i-1], neededTime[i])
            else:
                dp[i] = neededTime[i]

        return res
```

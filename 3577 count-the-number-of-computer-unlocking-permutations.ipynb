```python
class Solution:
    def countPermutations(self, complexity: List[int]) -> int:
        if complexity[0] != min(complexity) or complexity.count(complexity[0]) > 1:
            return 0
        
        def myFactorial(n):
            MOD = 10 ** 9 + 7

            res = 1
            for i in range(1, n+1):
                res = (res * i) % MOD
            
            return res
        
        return myFactorial(len(complexity) - 1)
```

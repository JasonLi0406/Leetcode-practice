```python
class Solution(object):
    def checkIfExist(self, arr):
        """
        :type arr: List[int]
        :rtype: bool
        """
        seen = set()

        for i in range(len(arr)):
            
            if arr[i] in seen:
                return True

            seen.add(2 * arr[i])
            seen.add(arr[i] // 2) if arr[i] % 2 == 0 else None

        return False
```

```python
arr = [0, -1, -3, 0]

Solution().checkIfExist(arr)
```

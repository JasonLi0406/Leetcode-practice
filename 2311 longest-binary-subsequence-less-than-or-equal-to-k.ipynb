```python
class Solution(object):
    def longestSubsequence(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
        def toDecimal(s):
            """
            :type s: str
            :rtype: int
            """
            n = len(s)
            i, res = 0, 0

            while i < n:
                if s[i] == '1':
                    res += 1 << (n-1-i)
                i += 1
            return res

        bitEqualOne = []

        n = len(s)
        for i in range(n):
            if s[i] == '1':
                bitEqualOne.append(n-1-i)
        
        value = toDecimal(s)

        i = 0
        while value > k:
            value -= 2 ** bitEqualOne[i]
            i += 1
            n -= 1

        return n
```

```python
from typing import List
import heapq
```

```python
class Solution:
    def mostBooked(self, n: int, meetings: List[List[int]]) -> int:
        meetings.sort()

        book_lst = [0] * n

        empty = []
        occupy = []

        for i in range(n):
            heapq.heappush(empty, i)

        for start, end in meetings:

            while occupy and occupy[0][0] <= start:
                _, room = heapq.heappop(occupy)
                heapq.heappush(empty, room)

            if empty:
                room = heapq.heappop(empty)
                heapq.heappush(occupy, (end, room))
                book_lst[room] += 1
            
            else:
                end_time, room = heapq.heappop(occupy)
                heapq.heappush(occupy, (end_time + end - start, room))
                book_lst[room] += 1

        return book_lst.index(max(book_lst))
```

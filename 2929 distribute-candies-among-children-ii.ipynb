```python
class Solution(object):
    def distributeCandies(self, n, limit):
        """
        :type n: int
        :type limit: int
        :rtype: int
        """
        def C(m, n):
            res = 1
            for i in range(1, n+1):
                res *= (m+1-i) / i
            return int(res)
        
        if 3*limit < n:
            return 0
        
        onePersonOverLimit = 0
        twoPersonOverLimit = 0

        if (limit+1) <= n:
            onePersonOverLimit = 3 * C(n+2-(limit+1), 2)
        if 2*(limit+1) <= n:
            twoPersonOverLimit = 3 * C(n+2-(limit+1)*2, 2)

        return C(n+2, 2) - onePersonOverLimit + twoPersonOverLimit
```

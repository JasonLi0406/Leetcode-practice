```python
from typing import List
from math import gcd
```

```python
class Solution:
    def minOperations(self, nums: List[int]) -> int:
        n = len(nums)

        if 1 in nums:
            return n - nums.count(1)

        if gcd(*nums) != 1:
            return -1

        minLength = n
        for i in range(n):
            temp = 0
            for j in range(i, n):
                temp = gcd(temp, nums[j])
                if temp == 1:
                    minLength = min(minLength, j - i + 1)
                    break
        return n + minLength - 2
```

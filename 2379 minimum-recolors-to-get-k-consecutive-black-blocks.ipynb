```python
class Solution(object):
    def minimumRecolors(self, blocks, k):
        """
        :type blocks: str
        :type k: int
        :rtype: int
        """
        cur, minimum = 0, k

        # Initialize the minimum recolored blocks
        for i in range(k):
            if blocks[i] == 'W':
                cur += 1

        minimum = cur

        for i in range(1, len(blocks)-k+1):
            if blocks[i+k-1] == 'W':
                cur += 1
            if blocks[i-1] == 'W':
                cur -= 1
            if cur < minimum:
                minimum = cur
        
        return minimum
        
```

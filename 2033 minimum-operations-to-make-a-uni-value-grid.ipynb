```python
def minOperations(grid, x):
    """
    :type grid: List[List[int]]
    :type x: int
    :rtype: int
    """
    m, n = len(grid), len(grid[0])

    if m == n == 1:
        return 0

    check = grid[0][0] % x

    lst = [0] * (m * n)

    for i in range(m):
        for j in range(n):
            if grid[i][j] % x != check:
                return False
            lst[i * n + j] = grid[i][j]

    lst.sort()

    median = lst[(m*n)//2]

    return [int(sum([abs(i - median)/x for i in lst])), median]
    
    
```

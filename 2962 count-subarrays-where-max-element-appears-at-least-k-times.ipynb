```python
class Solution(object):
    def countSubarrays(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        Max = max(nums)

        l, r, res = 0, 0, 0

        count = 0
        while r < len(nums):
            if nums[r] == Max:
                count += 1
            
            if count < k:
                r += 1
            else:
                res += len(nums) - r
                count = 0
                l += 1
                r = l
                
        return res
```

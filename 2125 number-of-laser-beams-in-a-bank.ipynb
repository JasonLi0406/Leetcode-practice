```python
class Solution:
    def numberOfBeams(self, bank: list[str]) -> int:

        def count(s: str) -> int:
            res = 0
            for i in s:
                res += 1 if i == '1' else 0
            return res
        
        dp = []

        for s in bank:
            if count(s) != 0:
                dp.append(count(s))

        if len(dp) == 1:
            return 0
        else:
            return sum([dp[i] * dp[i+1] for i in range(len(dp)-1)])
```

```python
class Solution:
    def numberOfBeams(self, bank: list[str]) -> int:

        prev, cur = 0, 0

        res = 0
        for s in bank:
            cur = s.count('1')
            if cur == 0:
                continue
            res += cur * prev
            prev = cur
        
        return res
```

```python
class Solution(object):
    def numberOfArrays(self, differences, lower, upper):
        """
        :type differences: List[int]
        :type lower: int
        :type upper: int
        :rtype: int
        """

        """
        # Version 1 -- Space Complexity -> O(n)
        hidden = [0] * (len(differences)+1)
        
        for i in range(1, len(hidden)):
            hidden[i] = hidden[i-1] + differences[i-1]
        
        ans = (upper - lower) - (max(hidden) - min(hidden)) + 1
        
        return ans if ans > 0  else 0
        """    

        # Version 2 -- Space Complexity -> O(1)
        Min, Max, Cur = 0, 0, 0
        
        for i in range(len(differences)):
            Cur += differences[i]
            if Cur > Max:
                Max = Cur
            elif Cur < Min:
                Min = Cur
        ans = (upper - lower) - (Max - Min) + 1
        
        return ans if ans > 0 else 0
```

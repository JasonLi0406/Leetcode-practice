```python
class Solution(object):
    def findSmallestInteger(self, nums, value):
        """
        :type nums: List[int]
        :type value: int
        :rtype: int
        """
        modulus_box = [0] * value
        
        for i in range(len(nums)):
            modulus_box[nums[i]%value] += 1

        least = min(modulus_box)
        idx = modulus_box.index(least)

        return least*value + idx
```

```python
def findSmallestInteger(nums, value):
    """
    :type nums: List[int]
    :type value: int
    :rtype: int
    """
    modulus_box = [0] * value
    for i in range(len(nums)):
        modulus_box[nums[i]%value] += 1

    least = min(modulus_box)
    idx = modulus_box.index(least)

    return least*value + idx
```

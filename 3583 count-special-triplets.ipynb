* **ChatGPT: Array version**

```python
class Solution:
    def specialTriplets(self, nums: List[int]) -> int:
        MOD = 10**9 + 7
        n = len(nums)

        # 值域最大 1e5，因此 freq array 比 dict 更快
        MAXV = 200000  # 需要到 2 * max(nums)
        freq_left = [0] * (MAXV + 1)
        freq_right = [0] * (MAXV + 1)

        # 初始化右邊頻率
        for x in nums:
            freq_right[x] += 1

        ans = 0

        for j, x in enumerate(nums):
            freq_right[x] -= 1   # j 不再屬於右邊
            
            target = x * 2
            if target <= MAXV:
                ans += freq_left[target] * freq_right[target]
                ans %= MOD
            
            freq_left[x] += 1   # j 變成左邊

        return ans % MOD
```

```python
def countSpecialTriplets(nums):
    MOD = 10**9 + 7
    n = len(nums)

    # 值域最大 1e5，因此 freq array 比 dict 更快
    MAXV = 200000  # 需要到 2 * max(nums)
    freq_left = [0] * (MAXV + 1)
    freq_right = [0] * (MAXV + 1)

    # 初始化右邊頻率
    for x in nums:
        freq_right[x] += 1

    ans = 0

    for j, x in enumerate(nums):
        freq_right[x] -= 1   # j 不再屬於右邊
        
        target = x * 2
        if target <= MAXV:
            ans += freq_left[target] * freq_right[target]
            ans %= MOD
        
        freq_left[x] += 1   # j 變成左邊

    return ans % MOD
```

* **ChatGPT: Dict version**

```python
from collections import Counter, defaultdict

class Solution:
    def specialTriplets(self, nums: List[int]) -> int:
        MOD = 10**9 + 7
        n = len(nums)

        freq_right = Counter(nums)     # 所有元素先放到右邊
        freq_left = defaultdict(int)

        ans = 0
        for x in nums:
            freq_right[x] -= 1
            target = x * 2

            # 直接用 dict 查詢（若不存在會回 0）
            ans = (ans + freq_left[target] * freq_right.get(target, 0)) % MOD

            freq_left[x] += 1

        return ans
```

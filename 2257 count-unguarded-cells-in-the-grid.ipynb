### **原始寫法**

你的程式邏輯正確、思路清晰，確實在小測資上可以得到正確結果 👍 \
不過會 TLE（Time Limit Exceeded） 的主因在於，你的寫法雖然看似是 O(m × n)，但實際上在「guard 眾多」或「grid 很大」時，掃描四個方向的 while loop 會導致「重複掃描同一區域」的問題，實際複雜度接近 O(G × (m + n))，其中 G 是守衛數量。 \
當 G 接近 5 × 10⁴，而 m、n 也很大時，就會超時。 



| 型別 | 每個元素的平均記憶體佔用（估算） | 存取速度 | 說明 |
| :---: | :---: | :---: | :--- |
| str (單字元) | 約 50–70 bytes | 較慢 | Python 的 str 是不可變（immutable）物件，長度為 1 也一樣建立完整物件。 |
| int (小整數) | 約 28 bytes（共用 small int pool） | 較快 | Python 有內建 small integer cache，速度明顯快於字串比較。 |
| numpy.int8（用 NumPy） | 1 byte | 最快 | 若用 NumPy array，速度接近 C 語言層級。 |

已將 str 表示法，改成數字表示法，先將 walls 和 guards 以同樣代號(2)表示，運算效能顯著提升

```python
from typing import List
class Solution:
    def countUnguarded(self, m: int, n: int, guards: List[List[int]], walls: List[List[int]]) -> int:
        grids = [[0] * n for _ in range(m)]

        for row, col in guards + walls:
            grids[row][col] = 2
        
        for row, col in guards:
            
            # Downwards
            i = row - 1
            while i >= 0:
                if grids[i][col] != 2:
                    grids[i][col] = 1
                    i -= 1
                else:
                    break
            
            # Upwards
            i = row + 1
            while i < m:
                if grids[i][col] != 2:
                    grids[i][col] = 1
                    i += 1
                else:
                    break

            # To the left
            i = col - 1
            while i >= 0:
                if grids[row][i] != 2:
                    grids[row][i] = 1
                    i -= 1
                else:
                    break

            # To the right
            i = col + 1
            while i < n:
                if grids[row][i] != 2:
                    grids[row][i] = 1
                    i += 1
                else:
                    break
        
        return sum(row.count(0) for row in grids)
```

### **LeetCode 上其他人的寫法**

| 項目 | 你的版本 | 她的版本 |
| :---: | :---: | :---: |
| 掃描方向 | 上下左右各掃一次 | 上下左右各掃一次 |
| 停止條件 | 遇到 'W' 或 'G' | 遇到 grid[x+dx][y+dy] == 2 |
| 計算 | unguarded | 統計 'x' 的數量	統計 0 的數量 | 

<br>

| 面向 | 你的版本 | 她的版本 |
| :---: | :---: | :---: |
|grid 型別	|str（'x','o','G','W'）|	int（0,1,2）|
|判斷邏輯	|多層 if + 多次比較	|單條 while 條件|
|每格狀態	|分成多種符號	|僅三種數字|
|視線阻擋判斷	|需判斷 'W' 或 'G'	|統一判斷 == 2|
|寫入次數	|可能多次寫 'o'	|多次寫 1 不影響效率|
|時間複雜度	|理論上 O(G × (m+n))，但實際常數極大	|同樣 O(G × (m+n))，但常數極小|
|結果	|可能 TLE	|可通過所有測資|

```python
class Solution:
    def countUnguarded(self, m: int, n: int, guards: List[List[int]], walls: List[List[int]]) -> int:
        grid = [[0] * n for _ in range(m)] 

        for x, y in guards + walls:
            grid[x][y] = 2  
        
        directions = [(-1, 0), (0, 1), (1, 0), (0, -1)]
        for gx, gy in guards:
            for dx, dy in directions:
                x, y = gx, gy
                while 0 <= x + dx < m and 0 <= y + dy < n and grid[x + dx][y + dy] != 2:
                    x += dx
                    y += dy
                    grid[x][y] = 1
        return sum(row.count(0) for row in grid)
```

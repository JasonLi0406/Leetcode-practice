```python
from collections import Counter
class Solution:
    def maxDifference(self, s: str) -> int:

        dic = Counter(s)
        oddMax, evenMin = 0, float('inf')
        for i in dic.values():
            if i % 2 == 0:
                if i < evenMin:
                    evenMin = i
            else:
                if i > oddMax:
                    oddMax = i
        return (oddMax - evenMin) # type: ignore
```

```python
class Solution:
    def countValidSelections(self, nums: list[int]) -> int:
        
        n = len(nums)
        prefix_sum_from_left = [0] * n
        prefix_sum_from_right = [0] * n

        for i in range(1, n):
            prefix_sum_from_left[i] = prefix_sum_from_left[i-1] + nums[i-1]
        
        for i in range(n-2, -1, -1):
            prefix_sum_from_right[i] = prefix_sum_from_right[i+1] + nums[i+1]

        count = 0
        for i in range(n):
            if nums[i] == 0:
                if prefix_sum_from_left[i] == prefix_sum_from_right[i]:
                    count += 2
                elif abs(prefix_sum_from_left[i] - prefix_sum_from_right[i]) == 1:
                    count += 1

        return count
```

```python
class Solution(object):
    def countHillValley(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        res = 0
    
        l, cur, r = 0, 1, 2
        while r < len(nums):
            if nums[r] == nums[cur]:
                r += 1
                continue
            if nums[cur] == nums[l]:
                cur += 1
                continue
            
            if (nums[cur] > nums[l]) and (nums[cur] > nums[r]):
                res += 1
            if (nums[cur] < nums[l]) and (nums[cur] < nums[r]):
                res += 1

            l = cur
            cur = r
            r += 1
        
        return res
```

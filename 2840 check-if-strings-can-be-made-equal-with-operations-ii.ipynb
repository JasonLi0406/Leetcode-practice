```python
class Solution:
    def checkStrings(self, s1: str, s2: str) -> bool:
        dic1_odd = dict()
        dic2_odd = dict()
        
        dic1_even = dict()
        dic2_even = dict()

        n = len(s1)

        for i in range(0, n, 2):
            if s1[i] in dic1_even:
                dic1_even[s1[i]] += 1
            else:
                dic1_even[s1[i]] = 1
            
            if s2[i] in dic2_even:
                dic2_even[s2[i]] += 1
            else:
                dic2_even[s2[i]] = 1

        for i in range(1, n, 2):
            if s1[i] in dic1_odd:
                dic1_odd[s1[i]] += 1
            else:
                dic1_odd[s1[i]] = 1
            
            if s2[i] in dic2_odd:
                dic2_odd[s2[i]] += 1
            else:
                dic2_odd[s2[i]] = 1

        return True if (dic1_even == dic2_even) and (dic1_odd == dic2_odd) else False
```

```python
class Solution:
    def bestClosingTime(self, customers: str) -> int:
        n = len(customers) + 1
        num_Y = customers.count('Y')
        num_N = n - num_Y

        post_Y = [num_Y] * n
        pre_N = [0] * n

        for i in range(1, n):
            if customers[i-1] == 'Y':
                post_Y[i] = post_Y[i-1] - 1
                pre_N[i] = pre_N[i-1]
            else:
                post_Y[i] = post_Y[i-1]
                pre_N[i] = pre_N[i-1] + 1
        
        penalty = [post_Y[i] + pre_N[i] for i in range(n)]

        return penalty.index(min(penalty))
```

```python
class Solution(object):
    def numberOfSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        a, b, c = list(), list(), list()

        for i in range(len(s)):
            if s[i] == 'a':
                a.append(i)
            elif s[i] == 'b':
                b.append(i)
            else:
                c.append(i)
        
        if (len(a) == 0) or (len(b) == 0) or (len(c) == 0):
            return 0
        
        a_pointer, b_pointer, c_pointer = 0, 0, 0
        count = 0

        while (a_pointer < len(a)) and (b_pointer < len(b)) and (c_pointer < len(c)):
            min_index = min(a[a_pointer], b[b_pointer], c[c_pointer])
            max_index = max(a[a_pointer], b[b_pointer], c[c_pointer])
            count += len(s) - max_index

            if s[min_index] == 'a':
                a_pointer += 1
            elif s[min_index] == 'b':
                b_pointer += 1
            else:
                c_pointer += 1
            
        return count
```

```python
class Solution(object):
    def numberOfSubstrings(self, s):
        """
        :type s: str
        :rtype: int
        """
        res = 0
        count = [-1] * 3
        for i, c in enumerate(s):
            count[ord(c) - 97] = i
            res += min(count) + 1
        return res
```

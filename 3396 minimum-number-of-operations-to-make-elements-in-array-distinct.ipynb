```python
class Solution(object):
    def minimumOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        
        hash_table = {}

        for i in nums:
            if i in hash_table:
                hash_table[i] += 1
            else:
                hash_table[i] = 1
        
        ans, i = 0, 0
        
        while i < len(nums):
            check_distinct = True
            for value in hash_table.values():
                if value > 1:
                    check_distinct = False
                    break

            if check_distinct:
                break
            
            try:
                hash_table[nums[i]] -= 1
                hash_table[nums[i+1]] -= 1
                hash_table[nums[i+2]] -= 1
            except:
                pass
            
            ans += 1
            i += 3
        
        return ans
```

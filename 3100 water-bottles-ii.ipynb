```python
class Solution(object):
    def maxBottlesDrunk(self, numBottles, numExchange):
        """
        :type numBottles: int
        :type numExchange: int
        :rtype: int
        """
        res = 0
        while numBottles >= numExchange:
            numBottles -= numExchange - 1
            res += numExchange
            numExchange += 1

        res += numBottles
        return res
```

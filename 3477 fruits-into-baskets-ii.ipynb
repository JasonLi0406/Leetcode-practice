```python
class Solution(object):
    def numOfUnplacedFruits(self, fruits, baskets):
        """
        :type fruits: List[int]
        :type baskets: List[int]
        :rtype: int
        """
        n = len(fruits)
        isEmpty = [True] * n

        for i in range(n):
            for j, (status, volume) in enumerate(zip(isEmpty, baskets)):
                if status and fruits[i] <= volume:
                    isEmpty[j] = False
                    break
                
        return sum(isEmpty) 
```

```python
def longestNiceSubarray(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    # Create two pointers:
    # Tail pointer is the end of the subarray
    # Cur pointer is the current number being checked
    # Count is the length of the subarray
    # Longest is the longest length found so far

    # Initialize variables
    storedNumber = nums[0]
    count, longest = 1, 1
    tail, cur = 0, 1
    count, longest = 1, 1

    # Initialize
    storedNumber = nums[0]
    
    while cur < len(nums):
        next = nums[cur]
        if next & storedNumber == 0:
            storedNumber = next | storedNumber
            count += 1
            cur += 1

            if count > longest:
                longest = count
        else:
            storedNumber = storedNumber ^ nums[tail]
            count -= 1
            tail += 1

    
    return longest
```

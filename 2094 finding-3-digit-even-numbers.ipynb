```python
from collections import Counter
class Solution(object):
    def findEvenNumbers(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        allEven = [100+2*i for i in range(450)]
        res = []

        checkDigit = Counter(digits)

        for i in allEven:
            dic = Counter(str(i))
            TF = True
            for digit, value in dic.items():
                if int(digit) not in checkDigit.keys():
                    TF = False
                    break
                if value > checkDigit[int(digit)]:
                    TF = False
                    break
            if TF:
                res.append(i)
        return res
```

```python
from collections import Counter
class Solution:
    def findEvenNumbers(self, dig: list[int]) -> list[int]:
        arr = []
        for i in range(100, 999, 2):
            flag = True
            seen = Counter(dig)
            for s in str(i):
                if int(s) not in seen:
                    flag = False
                    break
                else:
                    seen[int(s)] -= 1
                    if seen[int(s)] == 0:
                        del seen[int(s)]
            if flag:
                arr.append(i)
        
        return arr
```

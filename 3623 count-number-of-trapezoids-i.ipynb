```python
from typing import List
from collections import defaultdict
from math import comb
```

```python
class Solution:
    def countTrapezoids(self, points: List[List[int]]) -> int:
        y_axis = defaultdict(int)
        for i, j in points:
            y_axis[j] += 1
        
        valid = []
        for value in y_axis.values():
            if value >= 2:
                valid.append(comb(value, 2))
        
        return (sum(valid) ** 2 - sum([i ** 2 for i in valid])) // 2
```

```python
class Solution:
    def countTrapezoids(self, points: List[List[int]]) -> int:
        point_num = defaultdict(int)
        mod = 10**9 + 7
        ans, total_sum = 0, 0

        for point in points:
            point_num[point[1]] += 1

        for p_num in point_num.values():
            edge = p_num * (p_num - 1) // 2
            ans = (ans + edge * total_sum) % mod
            total_sum = (total_sum + edge) % mod
            
        return ans
```

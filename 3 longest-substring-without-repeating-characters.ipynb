```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        maximum = 0

        left, right = 0, 0
        while right < len(s):

            if s[right] not in s[left:right]:
                right += 1
                maximum = max(maximum, right - left)

            else:
                while s[right] in s[left:right]:
                    left += 1

        return maximum
```

```python
from typing import List
```

```python
class Solution:
    def maxDistance(self, nums1: List[int], nums2: List[int]) -> int:
        n = len(nums1)
        m = len(nums2)

        res = 0
        for i in range(n):

            # if nums1[i] > nums2[i]:
            #     continue

            target = nums1[i]

            l = i
            r = m-1
            while l <= r:
                mid = (l+r) // 2
                if nums2[mid] >= target:
                    l = mid + 1
                else:
                    r = mid - 1
            
            res = max(res, l-1-i)
        
        return res
```

* **Other solution (Two-pointer)**

```python
class Solution:
    def maxDistance(self, nums1: List[int], nums2: List[int]) -> int:
        i, j = 0, 0
        res = 0
        while i < len(nums1) and j < len(nums2):

            if nums2[j] >= nums1[i]: 
                j += 1
            else:
                i += 1
                j += 1
            # print(i, j)
            res = max(res, j - i)
        return max(res - 1, 0)
```

```python
nums1 = [55,30,5,4,2]
nums2 = [100,20,10,10,5]

Solution().maxDistance(nums1, nums2)
```

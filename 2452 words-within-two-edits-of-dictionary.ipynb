```python
from typing import List
```

```python
class Solution:
    def twoEditWords(self, queries: List[str], dictionary: List[str]) -> List[str]:
        res = []
        n = len(queries[0])
        for query in queries:
            flag = True
            for letter in dictionary:
                if flag:
                    countDiff = 0
                    for i in range(n):
                        if query[i] != letter[i]:
                            countDiff += 1
                            if countDiff >= 3:
                                break
                    if countDiff <= 2:
                        res.append(query)
                        flag = False
        
        return res
```

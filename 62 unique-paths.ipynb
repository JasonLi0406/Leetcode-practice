```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """
        
        # From math perspection
        # Ans = P(m-1+n-1, m-1)

        def P(a, b):
            nominator = 1
            denominator = 1

            for i in range(1, b+1):
                denominator *= i
                nominator *= (a + 1 - i)
            return nominator // denominator

        return P(m+n-2, m-1)
```

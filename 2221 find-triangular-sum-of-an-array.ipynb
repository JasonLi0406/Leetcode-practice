**Using Pascal's triangle**

```python
import math
def triangularSum(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    n = len(nums)
    
    res = 0
    for i in range(n):
        res += nums[i] * math.comb(n-1, i)
    
    return res % 10
```

```python
nums = [1,2,3,4,5]

triangularSum(nums)
```

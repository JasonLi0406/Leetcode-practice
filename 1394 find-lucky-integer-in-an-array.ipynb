```python
from collections import Counter
class Solution(object):
    def findLucky(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        dic = Counter(arr)

        res = -1
        for key, value in dic.items():
            if (key == value) and (key > res):
                res = key
        return res
```

```python
class Solution:
    def countCollisions(self, directions: str) -> int:
        collision = 0
        to_right = 0
        left_hand_side_exist_stop_car = False

        for i in directions:
            if i == 'R':
                to_right += 1
            
            elif i == 'S':
                left_hand_side_exist_stop_car = True
                if to_right:
                    collision += to_right
                    to_right = 0

            else:
                if to_right:
                    collision += (to_right + 1)
                    to_right = 0
                    left_hand_side_exist_stop_car = True
                elif left_hand_side_exist_stop_car:
                    collision += 1
        
        return collision
```

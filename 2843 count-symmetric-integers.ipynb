```python
class Solution(object):
    def countSymmetricIntegers(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: int
        """
        count = 0
        for n in range(low, high+1):
            digit = len(str(n))
            if digit%2==1:
                continue
            else:
                if sum([int(i) for i in str(n)[:digit//2]]) == sum([int(i) for i in str(n)[digit//2:]]):
                    count += 1
        
        return count
```

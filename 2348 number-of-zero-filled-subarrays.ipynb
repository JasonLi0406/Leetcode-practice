```python
class Solution(object):
    def zeroFilledSubarray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        zero_counter = []

        count = 0
        for number in nums:
            if number == 0:
                count += 1
            else:
                zero_counter.append(count)
                count = 0
        
        if count > 0:
            zero_counter.append(count)

        res = 0
        for count in zero_counter:
            res += (count * (count + 1)) // 2
        
        return res
```

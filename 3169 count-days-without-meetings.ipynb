```python
class Solution(object):
    def countDays(self, days, meetings):
        """
        :type days: int
        :type meetings: List[List[int]]
        :rtype: int
        """
        # Sort the meetings list by the start date and the end date
        meetings.sort(key=lambda x: [x[0], x[1]])

        # Create a list named meetings_duration
        meetings_duration = [meetings[0]]

        # Iterate through the meetings list and update meetings_duration list by merging the consecutive meeting date
        for meeting in meetings[1:]:
            if meeting[0] > meetings_duration[-1][-1] + 1:
                meetings_duration.append(meeting)
            else:
                meetings_duration[-1][-1] = max(meetings_duration[-1][-1], meeting[-1])
        
        # Calculate the total duration of the meetings
        total_duration = sum([(meeting[-1] - meeting[0] + 1) for meeting in meetings_duration])

        # Return the day without mmeeting
        return days - total_duration
```

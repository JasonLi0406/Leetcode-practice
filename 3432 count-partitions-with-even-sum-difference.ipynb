```python
from typing import List
```

```python
class Solution:
    def countPartitions(self, nums: List[int]) -> int:
        n = len(nums)
        total = sum(nums)

        return n - 1 if total % 2 == 0 else 0
```

```python
def countPartitions(nums: List[int]) -> int:
    n = len(nums)
    total = sum(nums)

    return n - 1 if total % 2 == 0 else 0
```

```python
nums = [10,10,3,7,6]

countPartitions(nums)
```

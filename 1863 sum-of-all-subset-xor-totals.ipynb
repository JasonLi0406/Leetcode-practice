```python
def subsetXORSum(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    
    Sum = 0

    for i in range(len(nums)):
        temp = 0

        for j in range(i, len(nums)):
            temp ^= nums[j]
            print(temp)
            Sum += temp
    
    return Sum
```

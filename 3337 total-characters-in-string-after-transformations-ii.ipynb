```python
class Solution(object):
    def lengthAfterTransformations(self, s, t, nums):
        """
        :type s: str
        :type t: int
        :type nums: List[int]
        :rtype: int
        """
        # print('nums=')
        # print(nums)
        # print()

        # Create the array
        pre = [0] * 26
        post = [0] * 26

        # Initialization
        for i in s:
            pre[ord(i)-ord('a')] += 1

        # print('pre=')
        # print(pre)
        # print()

        # Loop
        for i in range(t):
            for e in range(26):
                if pre[e] == 0:
                    continue
                if e + nums[e] < 26:
                    for j in range(e+1, e+nums[e]+1):
                        post[j] += pre[e]
                else:
                    for j in range(e+1, 26):
                        post[j] += pre[e]
                    for j in range(0, e+nums[e]-25):
                        post[j] += pre[e]
            # print()
            # print(f't={i+1}')
            # print(post)
            for e in range(26):
                pre[e] %= (10**9+7)

            pre = post[:]
            post = [0] * 26    
        
        return sum(pre) % (10**9+7)
```

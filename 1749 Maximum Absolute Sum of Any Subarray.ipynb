```python
class Solution(object):
    def maxAbsoluteSum(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        cur_p, cur_n = 0, 0
        record = [0, 0]
        
        for i in nums:
            # Find maximum value in the array
            cur_p += i if (cur_p+i >= 0) else -cur_p
            if cur_p >= record[0]:
                record[0] = cur_p

            # Find minimum value in the array
            cur_n += i if (cur_n+i <= 0) else -cur_n
            if cur_n <= record[1]:
                record[1] = cur_n
        
        return max(abs(record[0]), abs(record[1]))
```

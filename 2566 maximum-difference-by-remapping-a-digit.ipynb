```python
class Solution(object):
    def minMaxDifference(self, num):
        """
        :type num: int
        :rtype: int
        """
        def getMinMaxNumber(num, mode='min'):
            numString = str(num)
            if mode == 'min':
                fill_in = '0'
            else:
                fill_in = '9'

            check = ''
            for i in range(len(numString)):
                if numString[i] != fill_in:
                    check = numString[i]
                    break
            
            if len(check)==0:
                res = int(numString)
            else:
                res = int(''.join([i if i != check else fill_in for i in numString]))
            return res
            

        return getMinMaxNumber(num, 'max') - getMinMaxNumber(num, 'min')
```

```python
class Solution(object):
    def mergeArrays(self, nums1, nums2):
        """
        Merges two sorted 2D arrays based on ID using dictionaries, summing values for common IDs.

        :type nums1: List[List[int]]
        :type nums2: List[List[int]]
        :rtype: List[List[int]]
        """
        dict1 = {} # Dictionary to store id-value pairs from nums1
        dict2 = {} # Dictionary to store id-value pairs from nums2

        # Populate dictionaries from input arrays
        for id_val in nums1:
            dict1[id_val[0]] = id_val[1]
        for id_val in nums2:
            dict2[id_val[0]] = id_val[1]

        merged_ids = sorted(list(set(dict1.keys()) | set(dict2.keys()))) # Get sorted unique IDs from both dictionaries
        ans = [] # Initialize the list to store the merged result

        for id_val in merged_ids: # Iterate through each unique ID in sorted order
            val1 = dict1.get(id_val, 0) # Get value from dict1, default 0 if ID not present
            val2 = dict2.get(id_val, 0) # Get value from dict2, default 0 if ID not present
            ans.append([id_val, val1 + val2]) # Append the merged element with summed values to the result

        return ans # Return the merged and sorted array
```

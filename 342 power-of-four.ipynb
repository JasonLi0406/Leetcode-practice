```python
class Solution(object):
    def isPowerOfFour(self, n):
        """
        :type n: int
        :rtype: bool
        """
        if n < 0:
            return False

        return sum(map(int, bin(n)[2:])) == 1 and len(bin(n)[2:]) % 2 == 1
```

```python
class Solution(object):
    def getLongestSubsequence(self, words, groups):
        """
        :type words: List[str]
        :type groups: List[int]
        :rtype: List[str]
        """
        res = []

        check = not(groups[0])
        for i in range(len(words)):
            if groups[i] ^ check:
                res.append(words[i])
                check = groups[i]

        return res
```

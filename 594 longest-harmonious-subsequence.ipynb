```python
from collections import Counter
```

```python
class Solution(object):
    def findLHS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = Counter(nums)
        count = sorted(count.items())

        if len(count) == 1:
            return 0

        res = 0
        for i in range(len(count)-1):
            if count[i+1][0] - count[i][0] == 1:
                if count[i][1] + count[i+1][1] > res:
                    res = count[i][1] + count[i+1][1]
        return res
```

```python
class Solution(object):
    def findLHS(self, nums):
        from collections import Counter
        cnt=Counter(nums)
        res=0
        for k in cnt:
            if k+1 in cnt:
                res=max(res,cnt[k]+cnt[k+1])
        return res
```

```python
nums = [1,3,2,2,5,2,3,7]
Solution().findLHS(nums)
```

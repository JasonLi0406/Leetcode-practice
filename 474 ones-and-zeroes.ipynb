🧩 題目重點

我們要在 `strs` 這些二進位字串中，挑出最大數量的字串，使得：

所挑字串的「0 的總數」≤ `m`

所挑字串的「1 的總數」≤ `n`

---

💡 思路分析

這題的「容量」有兩個維度：

`m`: 可用的 0 數量

`n`: 可用的 1 數量

這讓問題變成「二維背包」：
每個字串就像一個「物品」，它消耗 `zeros` 個 0 和 `ones` 個 1，而它的「價值」是 1（代表挑這個字串可以讓集合數量 + 1）。

---

🔹 DP 定義

令 `dp[i][j]` 表示：

> 在最多使用 `i` 個 0、`j` 個 1 的情況下，能形成的最大子集合大小。

---

🔹 狀態轉移

對於每個字串 `(zeros, ones)`：

`dp[i][j] = max(dp[i][j], dp[i - zeros][j - ones] + 1)`

（前提是 i >= zeros 且 j >= ones）

注意更新順序：

必須 **從大到小更新 (反向 loop)**，以避免重複使用同一個字串。

---

🔹 初始化

`dp` 初始化為 0。

答案為 `dp[m][n]`。

---

🧠 範例推導

Example:

``` python
strs = ["10","0001","111001","1","0"]
m = 5, n = 3
```


每個字串的 (zeros, ones)：

``` python
"10"     -> (1,1)
"0001"   -> (3,1)
"111001" -> (2,4)
"1"      -> (0,1)
"0"      -> (1,0)
```


動態規劃計算後，最大能挑出 4 個字串。

---

✅ 程式碼（Python）

``` python
class Solution:
    def findMaxForm(self, strs: List[str], m: int, n: int) -> int:
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for s in strs:
            zeros = s.count('0')
            ones = s.count('1')

            # reverse loop to prevent reuse
            for i in range(m, zeros - 1, -1):
                for j in range(n, ones - 1, -1):
                    dp[i][j] = max(dp[i][j], dp[i - zeros][j - ones] + 1)

        return dp[m][n]
```

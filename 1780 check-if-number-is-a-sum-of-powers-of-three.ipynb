```python
class Solution(object):
    def checkPowersOfThree(self, n):
        """
        :type n: int
        :rtype: bool
        """
        while n >= 3:
            if (n % 3) == 2:
                return False
            else:
                n = (n - n%3) // 3
        
        if n == 2:
            return False
        else:
            return True
```

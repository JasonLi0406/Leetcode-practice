```python
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        res = []
        for level in range(1, numRows+1):
            res.append([1] * level)

            if numRows > 2:
                for i in range(1, level-1):
                    res[level-1][i] = res[level-2][i-1] + res[level-2][i]
        return res
```

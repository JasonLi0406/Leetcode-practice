```python
class Solution(object):
    def subarrayBitwiseORs(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        n = len(arr)
        seen = set()
        prev = set()
        
        for i in range(n):

            cur = {arr[i]}
            for j in prev:
                cur.add(arr[i] | j) 

            seen |= cur
            prev = cur
                
        return len(seen)
```

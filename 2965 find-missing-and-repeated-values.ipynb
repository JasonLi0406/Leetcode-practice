```python
def f(grid):
    a, b = 0, 0

    n = len(grid)
    check = [0] * n**2

    for i in range(n):
        for j in range(n):
            num = grid[i][j]
            if check[num-1] == 0:
                check[num-1] = 1
            else:
                a = num
                
    b = check.index(0) + 1
    return [a, b]
```

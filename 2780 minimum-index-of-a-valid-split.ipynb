```python
class Solution(object):
    def minimumIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Create a dictionary
        dic = dict()
        for i in nums:
            if i in dic:
                dic[i] += 1
            else:
                dic[i] = 1

        # Find the most frequent number and its frequency
        frequency = max(dic.values())
        most_frequent_number = [key for key, val in dic.items() if val == frequency][0]
        # print(most_frequent_number, frequency)

        # Replace the most frequent number with 1 and others with 0
        for i in range(len(nums)):
            nums[i] = 1 if nums[i] == most_frequent_number else 0
        # print(nums)
        
        # if len(nums) == 1:
        #     return 0

        # Check if there exists an index i such that the number of 1's in the two list is greater than the half of the list length
        left, right = 0, frequency
        for i in range(len(nums)):
            left += nums[i]
            right -= nums[i]

            if (left > (i+1)/2) and (right > (len(nums) - 1 - i)/2):
                return i
            
        # Otherwise, return -1
        return -1
```

```python
class Solution(object):
    def minimumIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        split_l = {}
        split_r = {}

        for n in nums:
            split_r[n] = split_r.get(n, 0) + 1
        
        for i in range(len(nums)):
            n = nums[i]
            split_l[n] = split_l.get(n, 0) + 1
            split_r[n] -= 1
            
            if split_l[n] * 2 > (i + 1) and split_r[n] * 2 > (len(nums) - i - 1):
                return i
            
        return -1
```

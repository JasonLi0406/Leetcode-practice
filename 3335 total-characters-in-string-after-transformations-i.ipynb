```python
class Solution(object):
    def lengthAfterTransformations(self, s, t):
        """
        :type s: str
        :type t: int
        :rtype: int
        """

        pre = [0] * 26

        # Initialization
        for i in s:
            pre[ord(i)-97] += 1

        post = pre.copy()
        
        for i in range(t):
            for e in range(25):
                post[e+1] = pre[e]
            post[0] = pre[-1]
            post[1] += pre[-1]

            pre = post.copy()

        return sum(pre) % (10**9+7)
            
```

```python
class Solution(object):
    def lengthAfterTransformations(self, s, t):
        """
        :type s: str
        :type t: int
        :rtype: int
        """

        dp = [0] * 26

        # Initialization
        for i in s:
            dp[ord(i)-97] += 1
        
        for i in range(t):
            last = dp.pop(-1)
            dp.insert(0, last)
            dp[1] += last

        return sum(dp) % (10**9+7)
            
```

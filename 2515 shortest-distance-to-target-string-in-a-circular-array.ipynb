```python
from typing import List
```

```python
class Solution:
    def closestTarget(self, words: List[str], target: str, startIndex: int) -> int:
        if target not in words:
            return -1
        
        n = len(words)

        leftwards = 0
        while words[startIndex-leftwards] != target:
            leftwards += 1
        
        rightwards = 0
        while words[(startIndex+rightwards)%n] != target:
            rightwards += 1

        return min(leftwards, rightwards)
```

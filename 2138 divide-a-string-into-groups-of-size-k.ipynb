```python
class Solution(object):
    def divideString(self, s, k, fill):
        """
        :type s: str
        :type k: int
        :type fill: str
        :rtype: List[str]
        """
        n = (len(s) - 1) // 3 + 1

        res = []
        for i in range(n-1):
            res.append(s[k*i:k*i+k])

        res.append(s[(n-1)*k:] + fill * (k-len(s[(n-1)*k:])))

        return res       
```

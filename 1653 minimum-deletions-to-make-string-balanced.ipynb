```python
from typing import List
from collections import Counter
```

```python
class Solution:
    def minimumDeletions(self, s: str) -> int:
        c = Counter(s)
        n = len(s)

        temp_a, temp_b = 0, 0
        for i in range(n):
            if temp_a < c['a']:
                if s[i] == 'a':
                    temp_a += 1
                else:
                    temp_b += 1
            else:
                break
        
        res = temp_b

        temp_a, temp_b = 0, 0
        for i in range(n-1, -1, -1):
            if temp_b < c['b']:
                if s[i] == 'b':
                    temp_b += 1
                else:
                    temp_a += 1
            else:
                break
        
        res = min(res, temp_a)

        return res
```

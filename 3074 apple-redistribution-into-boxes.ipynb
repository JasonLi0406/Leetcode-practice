```python
from typing import List
```

```python
class Solution:
    def minimumBoxes(self, apple: List[int], capacity: List[int]) -> int:
        capacity.sort(reverse=True)

        total = sum(apple)
        res = 0

        for i in range(len(capacity)):
            total -= capacity[i]
            res += 1

            if total <= 0:
                return res
        
        return -1
```

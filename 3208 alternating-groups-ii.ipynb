```python
class Solution(object):
    def numberOfAlternatingGroups(self, colors, k):
        """
        :type colors: List[int]
        :type k: int
        :rtype: int
        """
        colors = colors + colors[:k-1]

        count = 0
        switchcGroup = 1
        for i in range(1, len(colors)):
            if colors[i] != colors[i-1]:
                switchcGroup += 1
            else:
                if switchcGroup >= k:
                    count += (switchcGroup-k+1)
                switchcGroup = 1

        if switchcGroup >= k:
            count += (switchcGroup-k+1)

        return count
```

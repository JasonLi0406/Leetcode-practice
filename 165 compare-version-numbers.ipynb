```python
class Solution(object):
    def compareVersion(self, version1, version2):
        """
        :type version1: str
        :type version2: str
        :rtype: int
        """
        lst1 = version1.split('.')
        lst2 = version2.split('.')
        n1, n2 = len(lst1), len(lst2)

        for i in range(min(n1, n2)):
            if int(lst1[i]) > int(lst2[i]):
                return 1
            elif int(lst1[i]) < int(lst2[i]):
                return -1
        
        if n1 > n2:
            return 1 if sum([int(x) for x in lst1[n2:]]) > 0 else 0
        elif n1 < n2:
            return -1 if sum([int(x) for x in lst2[n1:]]) > 0 else 0
        else:
            return 0
```

```python
class Solution(object):
    def maxDistinctElements(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        nums.sort()

        cnt = 0
        prev = float("-inf")

        for num in nums:
            cur = min(max(num - k, prev + 1), num + k)

            if cur > prev:
                cnt += 1
                prev = cur
        
        return cnt
```

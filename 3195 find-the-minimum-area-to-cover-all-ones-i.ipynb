```python
class Solution(object):
    def minimumArea(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m = len(grid)
        n = len(grid[0])

        left, right, top, bottom = n-1, 0, m-1, 0

        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    left = min(left, j)
                    right = max(right, j)
                    top = min(top, i)
                    bottom = max(bottom, i)

        return (right - left + 1) * (bottom - top + 1)
```

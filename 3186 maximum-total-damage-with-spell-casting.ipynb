```python
from collections import defaultdict

class Solution(object):
    def maximumTotalDamage(self, power):
        """
        :type power: List[int]
        :rtype: int
        """
        spell = defaultdict(int)
        for i in power:
            spell[i] += 1
        spell = sorted(spell.items())

        n = len(spell)
        dp = [0] * n
        dp[0] = spell[0][0] * spell[0][1]
        for i in range(1, n):
            dp[i] = dp[i-1] + spell[i][0] * spell[i][1] if (spell[i][0] > spell[i-1][0] + 2) else max(dp[i-1], spell[i][0] * spell[i][1])
        
        return dp[-1]
```

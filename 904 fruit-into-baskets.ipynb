```python
from collections import defaultdict
class Solution(object):
    def totalFruit(self, fruits):
        """
        :type fruits: List[int]
        :rtype: int
        """
        n = len(fruits)

        collect, maxCollect = 0, 0
        left = 0
        seen = defaultdict(int)

        for right in range(n):

            seen[fruits[right]] += 1
            collect += 1

            while len(seen) > 2:
                seen[fruits[left]] -= 1
                collect -= 1
                if seen[fruits[left]] == 0:
                    del seen[fruits[left]]
                left += 1

            maxCollect = max(maxCollect, collect)    
        return maxCollect
```

```python
fruits = [1,2,3,2,2]
Solution().totalFruit(fruits)  # Output: 4
```

```python
from typing import List
```

```python
class Solution:
    def numMagicSquaresInside(self, grid: List[List[int]]) -> int:
        def check(i, j):
            # 取出 3x3
            a = grid[i-1][j-1:j+2]
            b = grid[i][j-1:j+2]
            c = grid[i+1][j-1:j+2]

            # row
            if sum(a) != 15 or sum(b) != 15 or sum(c) != 15:
                return False

            # column
            for k in range(3):
                if a[k] + b[k] + c[k] != 15:
                    return False

            # diagonals
            if a[0] + b[1] + c[2] != 15:
                return False
            if a[2] + b[1] + c[0] != 15:
                return False

            return True

        row = len(grid)
        col = len(grid[0])

        if row < 3 or col < 3:
            return 0
        
        res = 0
        for i in range(1, row - 1):
            for j in range(1, col - 1):
                if grid[i][j] == 5:

                    lst = []
                    lst.extend(grid[i-1][j-1:j+2])
                    lst.extend(grid[i][j-1:j+2])
                    lst.extend(grid[i+1][j-1:j+2])
                    
                    # 必須是 1~9
                    if set(lst) != set(range(1, 10)):
                        continue

                    if check(i, j):
                        res += 1

        return res
```

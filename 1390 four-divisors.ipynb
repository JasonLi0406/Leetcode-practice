```python
class Solution:
    def sumFourDivisors(self, nums: List[int]) -> int:
        N = max(nums)
        
        # build SPF
        spf = list(range(N + 1))
        for i in range(2, int(N ** 0.5) + 1):
            if spf[i] == i:
                for j in range(i * i, N + 1, i):
                    if spf[j] == j:
                        spf[j] = i

        ans = 0

        for x in nums:
            n = x
            factors = {}
            while n > 1:
                p = spf[n]
                factors[p] = factors.get(p, 0) + 1
                n //= p

            # case 1: p^3
            if len(factors) == 1:
                (p, e), = factors.items()
                if e == 3:
                    ans += 1 + p + p * p + x

            # case 2: p * q
            elif len(factors) == 2:
                (p, e1), (q, e2) = factors.items()
                if e1 == e2 == 1:
                    ans += 1 + p + q + x

        return ans
```

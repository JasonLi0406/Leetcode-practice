```python
from typing import List
```

```python
class Solution:
    def getDescentPeriods(self, prices: List[int]) -> int:
        prev = prices[0]
        count = 1
        res = 1

        for i in range(1, len(prices)):
            if prices[i] == prev - 1:
                count += 1
            else:
                count = 1

            prev = prices[i]
            res += count
            
        return res
```

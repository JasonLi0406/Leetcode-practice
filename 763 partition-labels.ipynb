```python
class Solution(object):
    def partitionLabels(self, s):
        """
        :type s: str
        :rtype: List[int]
        """
        container = dict()

        for i in range(len(s)):
            if s[i] in container:
                container[s[i]].append(i)
            else:
                container[s[i]] = [i]

        ans = []

        current = 0
        while len(container) > 0:

            include = {s[current]}
            tail = container[s[current]][-1]

            for element, lst in container.items():
                if lst[0] < tail:
                    include.add(element)
                    tail = max(lst[-1], tail)

            ans.append(tail + 1)

            current = (tail + 1)

            for i in include:
                container.pop(i)

        for i in range(len(ans)-1, 0, -1):
            ans[i] -= ans[i-1]

        return ans
```

```python
def partitionLabels(s):
    """
    :type s: str
    :rtype: List[int]
    """
    
    last_occurence = [0] * 26

    for i in range(len(s)):
        last_occurence[ord(s[i]) - ord('a')] = i
    
    return last_occurence
```

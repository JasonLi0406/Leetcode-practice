```python
class Solution(object):
    def countGoodNumbers(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        E = (n+1)//2
        O = n//2

        mod = 10**9 + 7
        def fastMod(x, n):
            if n == 1:
                return x
            if n == 0:
                return 1
            
            result = ((fastMod(x, n//2) % mod) ** 2) % mod
            if n%2 == 1:
                result = (result * x) % mod
            return result

        return (fastMod(5, E) * fastMod(4, O)) % mod
```

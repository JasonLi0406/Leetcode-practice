```python
class Solution(object):
    def findDiagonalOrder(self, mat):
        """
        :type mat: List[List[int]]
        :rtype: List[int]
        """
        m = len(mat)
        n = len(mat[0])

        res = [0] * (m * n)

        idx = 0
        for diagSum in range(m + n - 1):
            if diagSum % 2 == 0:
                r = min(diagSum, m - 1)
                c = diagSum - r
                while r >= 0 and c < n:
                    res[idx] = mat[r][c]
                    idx += 1
                    r -= 1
                    c += 1
            else:
                c = min(diagSum, n - 1)
                r = diagSum - c
                while c >= 0 and r < m:
                    res[idx] = mat[r][c]
                    idx += 1
                    r += 1
                    c -= 1

        return res
```

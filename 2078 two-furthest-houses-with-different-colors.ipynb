```python
from typing import List
```

```python
class Solution:
    def maxDistance(self, colors: List[int]) -> int:
        n = len(colors)
        d1, d2 = 0, 0

        for i in range(n-1, 0, -1):
            if colors[0] != colors[i]:
                d1 = i
                break

        for i in range(n-1):
            if colors[-1] != colors[i]:
                d2 = n - 1 - i
                break

        return max(d1, d2)
```

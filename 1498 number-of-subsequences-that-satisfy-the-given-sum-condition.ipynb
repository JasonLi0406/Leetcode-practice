```python
import math
class Solution(object):
    def numSubseq(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        nums.sort()

        n = len(nums)
        res = 0
        mod = 10**9 + 7

        left, right = 0, n - 1

        while left <= right:
            if nums[left] + nums[right] > target:
                right -= 1
            else:
                # print(left, right)
                res += pow(2, right - left, mod)
                left += 1

        return res
```

```python
def numSubseq(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: int
    """
    nums.sort()

    n = len(nums)
    res = 0

    right = n - 1
    for left in range(n):

        while left <= right:
            if nums[left] + nums[right] > target:
                right -= 1
            else:
                # print(left, right)
                res += 2 ** (right - left)
                break

    return res
```

```python
nums = [2,3,3,4,6,7]
target = 12

numSubseq(nums, target)
```

```python
def numSubseq(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: int
    """
    nums.sort()

    n = len(nums)
    res = 0

    left, right = 0, n - 1

    while left <= right:
        if nums[left] + nums[right] > target:
            right -= 1
        else:
            # print(left, right)
            res += 2 ** (right - left)
            left += 1

    return res
```

```python
class Solution(object):
    def makeFancyString(self, s):
        """
        :type s: str
        :rtype: str
        """
        res = ''

        temp = ''
        repeat = False
        for letter in s:
            if letter != temp:
                res += letter
                # Update condition
                temp = letter
                repeat = False

            else:
                if not(repeat):
                    res += letter
                    # Update condition
                    repeat = True
        
        return res
```

```python
class Solution(object):
    def makeFancyString(self, s):
        result = [s[0]]
        counter = 1
        string_length = len(s)

        if(string_length < 3):
            return s

        for i in range(1, string_length):
            if(s[i] != s[i-1]):
                counter = 1
            else:
                counter += 1

            if(counter < 3):
                result.append(s[i])

        return "".join(result)
```

```python
s = "leeetcode"

Solution().makeFancyString(s)
```

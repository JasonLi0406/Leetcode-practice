```python
class Solution(object):
    def minimumTeachings(self, n, languages, friendships):
        """
        :type n: int
        :type languages: List[List[int]]
        :type friendships: List[List[int]]
        :rtype: int
        """
        cannot_communicate = set()

        for (p1, p2) in friendships:
            p1_languages = set(languages[p1 - 1])
            p2_languages = set(languages[p2 - 1])

            if p1_languages.intersection(p2_languages):
                continue

            cannot_communicate.add(p1 - 1)
            cannot_communicate.add(p2 - 1)

        res = len(languages)
        
        for language in range(1, n+1):
            need_to_learn = 0

            for person in cannot_communicate:
                if language not in languages[person]:
                    need_to_learn += 1

            res = min(res, need_to_learn)
        
        return res
```

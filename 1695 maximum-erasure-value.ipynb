```python
class Solution(object):
    def maximumUniqueSubarray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        seen = set()
        left = 0
        total = 0
        max_score = 0

        for right in range(len(nums)):
            while nums[right] in seen:
                seen.remove(nums[left])
                total -= nums[left]
                left += 1
            seen.add(nums[right])
            total += nums[right]
            max_score = max(max_score, total)

        return max_score
```

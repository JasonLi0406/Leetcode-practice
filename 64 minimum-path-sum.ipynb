```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """
        m = len(grid[0])
        n = len(grid)

        dp = [0] * m

        dp[0] = grid[0][0]
        # Initialize the dp
        for i in range(1, m):
            dp[i] = dp[i-1] + grid[0][i]

        if n == 1:
            return dp[-1]

        else:
            # Downward the dp
            for i in range(1, n):
                dp[0] += grid[i][0]
                for j in range(1, m):
                    dp[j] = min(dp[j-1], dp[j]) + grid[i][j]
            return dp[-1] 
```

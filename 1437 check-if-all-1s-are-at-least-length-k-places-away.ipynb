```python
from typing import List
```

```python
class Solution:
    def kLengthApart(self, nums: List[int], k: int) -> bool:
        
        if nums.count(1) < 2: 
            return True 
            
        n = len(nums)
        begin = nums.index(1)
        space = 0

        for i in range(begin+1, n):
            if nums[i] == 0:
                space += 1
            else:
                if space < k:
                    return False
                space = 0
        
        return True
```

```python
class Solution:
    def kLengthApart(self, nums: List[int], k: int) -> bool:
        if k == 0:
            return True
        
        prev = None
        for i, num in enumerate(nums):
            if num == 1:
                if (prev is not None) and (i - prev <= k):
                    return False
                prev = i
        return True
```

```python
nums = [0,0,1,1,1]
k = 0

Solution().kLengthApart(nums, k)
```

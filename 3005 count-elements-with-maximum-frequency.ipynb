```python
from collections import Counter
class Solution(object):
    def maxFrequencyElements(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        lst = Counter(nums)
        max_freq = max(lst.values())

        res = 0
        for num, freq in lst.items():
            if freq == max_freq:
                res += freq

        return res
```

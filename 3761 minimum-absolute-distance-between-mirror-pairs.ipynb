```python
from typing import List
```

```python
class Solution:
    def minMirrorPairDistance(self, nums: List[int]) -> int:
        seen = dict()
    
        res = float('inf')
        for i, num in enumerate(nums):
            if num in seen.keys():
                res = min(res, i - seen[num])
            
            seen[int(str(num)[::-1])] = i

        return -1 if res == float('inf') else res
```

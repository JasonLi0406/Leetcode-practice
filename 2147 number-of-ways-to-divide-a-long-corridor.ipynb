```python
class Solution:
    def numberOfWays(self, corridor: str) -> int:
        S = corridor.count('S')

        if S % 2 == 1 or S == 0:
            return 0
        
        MOD = 10 ** 9 + 7
        cur_S = 0
        cur_P = 1
        valid = []

        corridor = corridor.strip('P')

        for i in corridor:
            if i == 'S':
                if cur_S == 2:
                    valid.append(cur_P)
                    cur_P = 1
                    cur_S = 1
                else:
                    cur_S += 1
            else:
                if cur_S == 2:
                    cur_P += 1
        
        res = 1
        for i in valid:
            res = res * i % MOD
        
        return res
```

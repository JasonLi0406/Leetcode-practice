```python
class Solution(object):
    def findKDistantIndices(self, nums, key, k):
        """
        :type nums: List[int]
        :type key: int
        :type k: int
        :rtype: List[int]
        """
        res = set()
    
        i = 0
        n = len(nums)
        while i < n:
            if nums[i] != key:
                i += 1
            else:
                for j in range(i-k, i+k+1):
                    if (j >= 0) and (j < n):
                        res.add(j)
                i += k

        return sorted(list(res))
```

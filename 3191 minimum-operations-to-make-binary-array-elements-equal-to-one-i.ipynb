```python
class Solution(object):
    def minOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Define the 'flip' operation
        def flip(n):
            return 1 if n == 0 else 0
        
        count = 0
        for i in range(len(nums)-2):
            if nums[i] == 0:
                for j in range(3):
                    nums[i+j] = flip(nums[i+j])
                count += 1
            # print(nums)

        return count if (nums[-1] and nums[-2]) else -1
```

```python
def minOperations(nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    
    count = 0
    for i in range(len(nums)-2):
        if nums[i] == 0:
            for j in range(3):
                nums[i+j] ^= 1 
            count += 1
        # print(nums)

    
    return count if (nums[-1] and nums[-2]) else -1
```

```python
nums = [0,1,1,1]
minOperations(nums)
```

```python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        dic = {0: 0, 1: 0, 2: 0}
        for i in nums:
            dic[i] += 1
        
        for i in range(len(nums)):
            if i < dic[0]:
                nums[i] = 0
            elif (dic[0] <= i) and (i < dic[0]+dic[1]):
                nums[i] = 1
            else:
                nums[i] = 2
        
        return nums
```

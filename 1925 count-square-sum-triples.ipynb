```python
class Solution:
    def countTriples(self, n: int) -> int:
        res = 0
        for c in range(5, n + 1):
            for a in range(1, c):
                for b in range(1, a):
                    if (a ** 2 + b ** 2) > (c ** 2):
                        break
                    elif (a ** 2 + b ** 2) == (c ** 2):
                        res += 2
                        break
        return res
```

* **ChatGPT: 優化**

```python
class Solution:
    def countTriples(self, n: int) -> int:
        res = 0

        for a in range(1, n):
            a2 = a * a
            for b in range(a + 1, n):
                c2 = a2 + b * b
                c = math.isqrt(c2)

                if c > n:
                    break

                if c * c == c2:
                    res += 2
                    
        return res
```

```python
n = 250

Solution().countTriples(n)
```

* **ChatGPT: Euclid 法**

```python
class Solution:
    def countTriples(self, n: int) -> int:
        res = 0
        
        # 最大 m 只要滿足 m^2 + 1 <= n
        m_max = int(math.sqrt(n))

        for m in range(2, m_max + 1):
            for s in range(1, m):  # s = n in Euclid formula
                # 必須互質 & 一奇一偶
                if (m - s) % 2 == 1 and math.gcd(m, s) == 1:

                    # primitive triple
                    a = m * m - s * s
                    b = 2 * m * s
                    c = m * m + s * s

                    if c > n:
                        break

                    # scale multiplier
                    k = 1
                    while k * c <= n:
                        A = k * a
                        B = k * b
                        C = k * c

                        if A <= n and B <= n and C <= n:
                            res += 2  # (a,b,c) & (b,a,c)

                        k += 1
        return res
```

```python
n = 250

Solution().countTriples(n)
```

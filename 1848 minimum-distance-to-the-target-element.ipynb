```python
from typing import List
```

```python
class Solution:
    def getMinDistance(self, nums: List[int], target: int, start: int) -> int:
        n = len(nums)
        for i in range(n):
            if 0 <= start + i <= n - 1 and nums[start+i] == target:
                return i
            elif 0 <= start - i <= n - 1 and nums[start-i] == target:
                return i
        return -1
```

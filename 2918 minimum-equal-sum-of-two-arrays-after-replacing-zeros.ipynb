```python
class Solution(object):
    def minSum(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: int
        """
        o1, o2 = 0, 0
        sum1, sum2 = 0, 0

        for i in nums1:
            if i == 0:
                sum1 += 1
                o1 += 1
            else:
                sum1 += i
        
        for i in nums2:
            if i == 0:
                sum2 += 1
                o2 += 1
            else:
                sum2 += i

        if (o1 == 0) and (sum1 < sum2):
            return -1
        elif (o2 == 0) and (sum2 < sum1):
            return -1
        else:
            return max(sum1, sum2)
```

```python
class Solution(object):
    def applyOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        for i in range(len(nums)-1):
            if nums[i] == nums[i+1]:
                nums[i] *= 2
                nums[i+1] = 0

        ans = [0] * len(nums)
        pnt = 0
        for i in range(len(nums)):
            if nums[i] != 0:
                ans[pnt] = nums[i]
                pnt += 1

        return ans
```

```python
nums = [1,2,2,1,1,0]

print(Solution().applyOperations(nums)) # Output: [1, 4, 2, 0, 0]
```
